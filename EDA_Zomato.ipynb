Merging data from another file

```python
df_country=pd.read_excel('Country-Code.xlsx')

df_country.head()
```

```python
#MERGE COUNTRY TO THE ZOMATO DATASET


final_df=pd.merge(df,df_country,on='Country Code',how='left')

final_df.columns #Country as a col has been added to the new dataframe


```

```python
#checking name of countries

country_name=final_df.Country.value_counts().index

country_name
```

```python
#checking values assocoated with countries

final_df.Country.value_counts()
```

```python
#taking values to plot a pie chart

country_val=final_df.Country.value_counts().values

country_val
```

CREATING PIE CHART

```python
#CREATING PIE CHART FOR TOP THREE COUNTRY THAT USE ZOMATO

plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
plt
```

```python
#converting the rating table to a dataframe
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index ()

#WE CAN SEE THAT THERE IS A COLOUMN NAMED ZERO WHICH BASICALLY HAS THE RATING COUNT, THEREFORE WE NEED TO CHANGE THE NAME FROM ZERO TO RATING COUNT
```

```python
#WE CAN SEE THAT THERE IS A COLOUMN NAMED ZERO WHICH BASICALLY HAS THE RATING COUNT, THEREFORE WE NEED TO CHANGE THE NAME FROM ZERO TO RATING COUNT

ratings=final_df.groupby([ 'Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

OBSERVATION FORM RATING TABLE:

1> When Rating is Between 4.5 to 4.9----->Excellent 

2> When Rating is Between 4.0 to 4.4----->Very good

3> When Rating is Between 3.5 to 3.9----->Good 

4> When Rating is Between 3.0 to 3.4----->Average

5> When Rating is Between 2.5 to 2.9----->Average 

6> When Rating is Between 2.0 to 2.4----->Poor





```python
#Barplot

sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)

#increase dimensions of the plot

matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings)


#when we plot using the rating color we can see that the color doesn't matches the description therefor we need to map the colors, we will be using the feature palette
```

```python
#plotting with correct  rating colors 
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

OBSERVATION:

1> Not rated count is very high

2> Maximun number of ratings are between 2.5 to 3.4

```python
# count plot----> plots based on category

sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
